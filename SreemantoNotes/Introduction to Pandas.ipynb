```python
earthquake_filt['Lat'] = earthquake_filt['Latitude']
```

```python
def checkfunc(x):
    if x >= 23.6:
        return 'North of Tropics'
    elif x <=-23.6:
        return 'South of Tropics'
    else:
        return "Tropics"
```

```python
earthquake_filt['Region'] = earthquake_filt['Lat'].apply(lambda x : checkfunc(x))
earthquake_filt.head(5)
```

```python
figure = plt.figure(figsize=(15,8))
plt.title("Region-wise count of earthquakes")
earthquake_filt['Region'].value_counts().sort_index().plot(kind='bar');
```

```python
#From the figure above we see that the tropics i.e. the region between the Tropic of Cancer and the Tropic of Capricorn see the most number of earthquakes.
#Possibly being closer to the equator there may be more seismological activity due to some geological reason
#Also more earthquakes happen north of the tropics than south - possibly as there is more landmass
```

```python
earthquake_filt['Long'] = earthquake_filt['Longitude']
```

```python
def checkfunc2(x):
    if x >= 0 and x <= 180 :
        return 'Eastern Hemisphere'
    else:
        return "Western Hemisphere"
```

```python
earthquake_filt['Hemisphere'] = earthquake_filt['Long'].apply(lambda x : checkfunc2(x))
earthquake_filt.head(5)
```

```python
figure = plt.figure(figsize=(15,8))
plt.title("Hemisphere-wise count of earthquakes")
earthquake_filt['Hemisphere'].value_counts().sort_index().plot(kind='bar');
```

```python
#Seems the eastern hemisphere has more earthquakes. This dismissed a pre conceived notion that most earthquakes happen in the western pacific region
```

```python
#Filtering out the datafram for earthquake greater than 7 magnitude
earthquake_filt7 = earthquake_filt[earthquake_filt.Magnitude > 7]

# we need the maximum and minimum coordinates as a part of syntax for drawing the bounding box.
# Also we need min and max value to extract the map use this link - https://www.openstreetmap.org/export#map=1/26/-88
BBox = ((earthquake_filt7.Longitude.min(),   earthquake_filt7.Longitude.max(),      
         earthquake_filt7.Latitude.min(), earthquake_filt7.Latitude.max()))
```
