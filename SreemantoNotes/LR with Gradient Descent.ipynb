```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([2,4,5])
y=np.array([1.2,2.8,5.3])
```

```python
plt.scatter(X,y)
plt.show()
```

```python
#initialize parameters
m = 1  #slope
c = 0  #intercept
learning_rate = 0.001  # step size
iterations = 1000
error = []
```

```python
for its in range(iterations):
    loss = 0
    cost_m = 0
    cost_c = 0
    
    for i in range(len(X)):
        
        y_pred = (m*X[i]  + c)     #predicted values
        
        loss = loss + (y[i]-y_pred)**2  #calculating total loss for each ierations
        
        for j in range(len(X)): # calculating derivates one at a time for each point
            
            partial_dev_wrt_c = -2*(y[j] - (m*X[j]+c))                   #partial derivate
            
            partial_dev_wrt_m = (-2*X[j]*(y[j] -  (m*X[j]+c)))           #partial derivate
            
             
            cost_c = cost_c + partial_dev_wrt_c   
            cost_m = cost_m + partial_dev_wrt_m
            
        c = c - cost_c*learning_rate   #update values
        m = m - cost_m*learning_rate   #update values
        
    error.append(loss)
```

```python
plt.figure(figsize=(20,10))
plt.plot(np.arange(1,len(error)+1),error,color='red',linewidth=5)
plt.title('Iteration vs Error')
plt.xlabel("Iterations")
plt.ylabel("Error")
```
