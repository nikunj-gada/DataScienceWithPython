### Creating 'day_bucket' column

```python
df2['day_bucket']=(df2['date']-pd.datetime(2020,10,22)).astype(str).str[0:2]
```

```python
df2=df2.drop(['date'],axis=1)
```

```python
df2
```

```python
df2['day_bucket']=df2['day_bucket'].map({  '-9' : '-1 to -10',
                                           '-8' : '-1 to -10',
                                           '-7' : '-1 to -10',
                                           '-6' : '-1 to -10',
                                           '-5' : '-1 to -10',
                                           '-4' : '-1 to -10',
                                           '-3' : '-1 to -10',
                                           '-2' : '-1 to -10',
                                           '-1' : '-1 to -10',
                        '0 ' : '0 to 2',
                        '1 ' : '0 to 2',
                        '2 ' : '0 to 2',
            '3 ' : '3 to 7',
            '4 ' : '3 to 7',
            '5 ' : '3 to 7',
            '6 ' : '3 to 7',
            '7 ' : '3 to 7'
                          })
```

```python
df2
```

```python
df2[df2["day_bucket"] == '0 to 2'].groupby(['format','division_name','day_bucket'])['orderval','rcvval'].sum()
```

```python
df2[df2["day_bucket"] == '-1 to -10'].groupby(['format','division_name'])['orderval','rcvval'].sum()
```

```python
d1 = df2[df2["day_bucket"] == '-1 to -10'].groupby(['format','division_name'])['orderval','rcvval'].sum()
d2 = df2[df2["day_bucket"] == '0 to 2'].groupby(['format','division_name'])['orderval','rcvval'].sum()
d3 = df2[df2["day_bucket"] == '3 to 7'].groupby(['format','division_name'])['orderval','rcvval'].sum()
```

```python
d1['fillrate']=d1['rcvval']/d1['orderval']
d2['fillrate']=d2['rcvval']/d2['orderval']
d3['fillrate']=d3['rcvval']/d3['orderval']
```

```python
dflist=[d1,d2,d3]
df_final=pd.concat(dflist,axis=1)
df_final
```

```python
d1.columns = ['orderval_B1', 'rcvval_B1', 'fillrate_B1']
d2.columns = ['orderval_B2', 'rcvval_B2', 'fillrate_B2']
d3.columns = ['orderval_B3', 'rcvval_B3', 'fillrate_B3']
```

```python
dflist=[d1,d2,d3]
df_final=pd.concat(dflist,axis=1)
df_final
```
