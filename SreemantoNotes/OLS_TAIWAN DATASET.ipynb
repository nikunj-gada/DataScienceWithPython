```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
data = pd.read_csv('taiwan_housing_data.csv')
data
```

```python
X = data[['house_age', 'dist_MRT', 'num_stores', 'Xs', 'Ys', 'dist_fr_cen']]
y = data.price

#X = sm.add_constant(X)
```

```python
X = sm.add_constant(X)
```

```python
s_model = sm.OLS(y, X)
s_results = s_model.fit()
s_results.summary()
```

```python
X_new = data[['house_age', 'dist_MRT', 'num_stores', 'Xs', 'Ys','dist_fr_cen']]
X_new['dist_MRT'] = np.log(X_new.dist_MRT) #log scaling
#print(X_new['dist_MRT'])
y = data.price

X = sm.add_constant(X_new)

s_model = sm.OLS(y, X)
s_results = s_model.fit()
s_results.summary()
```

# Treating Multicolinearity

```python
# #m1x1 + m2X2 +m3x3
# x1 = m12X2 + m13X3 => big R2 value if they show linear relationship (1/1-R2) => higher VIF value

# x2 = mX1 + mX3 => big R2 value
```

```python
# Checking VIF Values
print ("\nVariance Inflation Factor")

cnames = X_new.columns

for i in np.arange(0,len(cnames)):

    xvars = list(cnames)
    
    
    #print(xvars)
    
    yvar = xvars.pop(i)
    
    #print(yvar)
    
    mod = sm.OLS(X_new[yvar],(X_new[xvars]))
    
    res = mod.fit()
    
    vif = 1/(1-res.rsquared)
    print(yvar,'------>',round(vif,3))
```

#### we see that dist_MRT,dist_fr_cen show high multicollinearity we can do without including them lets see how it goes

```python
#discarding dist_MRT and the re-modelling
X2 = X_new[['house_age', 'num_stores', 'Xs', 'Ys','dist_fr_cen']]
X = sm.add_constant(X2)

s_model = sm.OLS(y, X2)
s_results = s_model.fit()
s_results.summary()
```

#### Sharp increase in R-Square and adjusted R-Square is visible meaning our model has fitted well. Now Let drop dist_fr_cen and check our results4

```python
X3 = X_new[['house_age','dist_MRT', 'num_stores', 'Xs', 'Ys']]
X = sm.add_constant(X3)

s_model = sm.OLS(y, X3)
s_results = s_model.fit()
s_results.summary()
```

#### Removing dist_fr_cen has even increased our r-squared value even more.

```python
# Check VIF values now
print ("\nVariance Inflation Factor")


cnames = X3.columns

for i in np.arange(0,len(cnames)):
    xvars = list(cnames)
    #print(xvars)
    yvar = xvars.pop(i)
    #print(yvar)
    mod = sm.OLS(X3[yvar],(X3[xvars]))
    res = mod.fit()
    
    vif = 1/(1-res.rsquared)
    print(yvar,'------>',round(vif,3))
```

#### We con see a significant drop in VIF values for all the feature indicating very low multicollinearity. Lets Check this by having a look at the correlaton matrix which is used to see correlation between features

```python
y = y.drop(270)
X3 = X3.drop(index=270)

s_model = sm.OLS(y, X3)
s_results = s_model.fit()
s_results.summary()
```

We see that there R-square value jumps by 1% after removal of the outlier point.

In conclusion we started with the base model which gave us an R-Square value of 57%.

We made 3 transformations.

1. Taking log of the features which had higher values.
2. Addressed the multicolinearity problem by checking the VIF values and the removed one of the correalted features.
3. Also removed an outlier point.

In doing so our model improved greatly and fit well very well with the given dataset.
