```python
import pandas as pd
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
```

```python
df = pd.read_csv('C:/Users/skesh/Imarticus/Datasets/Churn.csv')
df.head(5)
```

```python
df['Churn'] = df['Churn'].replace({'Yes':1,'No':0})
df.head(5)
```

```python
cols = ['gender', 'tenure', 'PhoneService', 'Contract', 'PaperlessBilling',
        'MonthlyCharges',]
```

```python
X = df[cols]
y = df['Churn']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
scale = StandardScaler()
ohe = OneHotEncoder()
logreg = LogisticRegression()
```

```python
transformer = make_column_transformer(
    (ohe,['gender','PhoneService', 'Contract', 'PaperlessBilling']),
    (scale,['tenure','MonthlyCharges']),
    remainder='passthrough'    
)
```

```python
pipe = make_pipeline(transformer,logreg)
pipe.fit(X_train,y_train)
```

```python
df = pd.read_csv('http://bit.ly/kaggletrain')
df = df.loc[df.Embarked.notna(), ['Survived', 'Pclass', 'Sex', 'Embarked']]
X = df.drop('Survived', axis='columns')
y = df.Survived
```

```python
column_trans = make_column_transformer(
    (OneHotEncoder(), ['Sex', 'Embarked']),
    remainder='passthrough')
logreg = LogisticRegression()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
pipe = make_pipeline(column_trans, logreg)
cross_val_score(pipe, X, y, cv=10, scoring='accuracy').mean()
```

```python
X_new = X.sample(5, random_state=99)
X_new
```

```python
pipe.fit(X, y)
pipe.predict(X_new)
```
