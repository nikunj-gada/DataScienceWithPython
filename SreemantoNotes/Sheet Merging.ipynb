```python
import pandas as pd
```

```python
df = pd.DataFrame()
for i in range(0,5):
    #print(i)
    df1 = pd.read_excel('sample_file.xlsx',sheet_name=i)
    df1 = df1.set_index(['User Name','First Name','Last Name','User Lock Status']).stack().reset_index().rename(
        columns={'level_4':'Categroy',0:'Score'})
    df = pd.concat([df,df1])    
```
