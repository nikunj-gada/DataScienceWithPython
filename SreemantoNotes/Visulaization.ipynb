```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cn = pd.read_csv('https://raw.githubusercontent.com/BadreeshShetty/Data-Visualization-using-Matplotlib/master/datasets/countries.csv')
```

```python
cn.to_csv('countries.csv')
```

```python
data_2007 = cn[cn['year']==2007]
```

## Scatter Plot

```python
plt.scatter(data_2007['gdpPerCapita'],data_2007['lifeExpectancy'],s=30,c='brown')
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
plt.scatter(np.log10(data_2007['gdpPerCapita']),data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
```

## Bar plot

```python
top_10_p=data_2007.sort_values('population',ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))

x=range(10)
plt.bar(x,top_10_p['population'])
plt.xticks(x,top_10_p['country'])
plt.xlabel('Countries')
plt.ylabel('Population in Billions')
plt.title('10 Most Populous Countries in 2007')
plt.show()
```

## Line plots

```python
stock_data  = pd.read_csv('https://raw.githubusercontent.com/BadreeshShetty/Data-Visualization-using-Matplotlib/master/datasets/stocks.csv')
```

```python
stock_data.to_csv('Stocks.csv',index=False)
```

```python
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data.head()
```

```python
fig = plt.figure(figsize=(10,6))

plt.plot(stock_data['Date'],
         stock_data['AAPL'],
         color='green')
```

```python
fig = plt.figure(figsize=(10,6))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])

ax1.set_title('AAPL vs IBM(inset)')

ax1.plot(stock_data['Date'],
         stock_data['AAPL'],
         color='green')

ax2.plot(stock_data['Date'],
         stock_data['IBM'], 
         color='blue')
plt.show()
```

#### subplots

```python
# Figsize for width and height of plot
fig = plt.figure(figsize=(15,7))

fig.suptitle('Stock price comparison 2007-2017', 
             fontsize=20)

ax1 = fig.add_subplot(221)
ax1.set_title('MSFT')

ax1.plot(stock_data['Date'],
         stock_data['MSFT'], 
         color='green')

ax2 = fig.add_subplot(222)
ax2.set_title('GOOG')

ax2.plot(stock_data['Date'],
         stock_data['GOOG'], 
         color='purple')

ax3 = fig.add_subplot(223)
ax3.set_title('SBUX')

ax3.plot(stock_data['Date'],
         stock_data['SBUX'], 
         color='magenta')

ax4 = fig.add_subplot(224)
ax4.set_title('ADBE')

ax4.plot(stock_data['Date'],
         stock_data['ADBE'], 
         color='orange')

plt.show()
```

## Stack plots

```python
x = np_data['Year']
```

```python
y = np.vstack([np_data['Badlands'], 
               np_data['GrandCanyon'], 
               np_data['BryceCanyon']])
```

```python
y
```

```python
# Labels for each stack
plt.figure(figsize=(15,7))
labels = ['Badlands', 'GrandCanyon', 'BryceCanyon']

# Colors for each stack
colors = ['brown', 'tomato', 'skyblue']

#  Similar to pandas df.plot.area()
plt.stackplot(x, y, labels=labels,
              colors=colors, edgecolor='black')

# Plots legend to the upperleft of Figure
plt.legend(loc=2)

plt.show()
```

## Stem Plot

```python
np_stem = np_data.copy()
```

```python
np_stem[['Badlands', 'GrandCanyon','BryceCanyon']] = np_data[['Badlands', 'GrandCanyon','BryceCanyon']].diff()
np_stem
```

```python
plt.figure(figsize=(15,11))

plt.suptitle('Change in Number of Visitors', y=0.94)

plt.subplot(311)
plt.stem(np_stem['Year'],
         np_stem['Badlands'],)
#          markerfmt = 'b_',
#          linefmt = 'r--',
#          basefmt = 'g:')
plt.title('GrandCanyon')

plt.subplot(312)
plt.stem(np_stem['Year'],
         np_stem['GrandCanyon'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:')
plt.title('Badlands')

plt.subplot(313)
plt.stem(np_stem['Year'],
         np_stem['BryceCanyon'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:')
plt.title('BryceCanyon')

plt.show()
```
